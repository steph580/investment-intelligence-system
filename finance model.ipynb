```python

import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format

import yfinance as yf
def analyze_company(ticker_symbol):

    #profit margin
    company = yf.Ticker(ticker_symbol)
    income_statement = company.income_stmt

    revenue_generated = income_statement.loc["Total Revenue"]
    profit_generated = income_statement.loc["Net Income"]

    profit_margin_calculation = (profit_generated / revenue_generated) * 100
    print("Profit Margin by year:")
    print((profit_margin_calculation).round(2))

    #current ratio
    balance_sheet_statement = company.balance_sheet
    current_assets_generated = balance_sheet_statement.loc["Current Assets"]
    current_liabilities_generated = balance_sheet_statement.loc["Current Liabilities"]

    current_ratio_calculation = (current_assets_generated / current_liabilities_generated)
    print(f"Current ratio")
    print((current_ratio_calculation).round(2))

    #debt repayment ratio
    debt_generated = balance_sheet_statement.loc["Total Debt"]

    print("Total Debt by year:")
    print(debt_generated.round(2))

    debt_repayment_ratio_calculation = profit_generated / debt_generated
    print(f"debt repayment ratio")
    print((debt_repayment_ratio_calculation).round(2))

    #return on equity
    equity_generated = balance_sheet_statement.loc["Stockholders Equity"]
    roe = (profit_generated / equity_generated) * 100

    print("ROE")
    print((roe).round(2))

    #debt to equity
    debt_to_equity = debt_generated/equity_generated
    print(f"Debt to Equity Ratio")
    print((debt_to_equity).round(2))

    #free cash flow
    cash_flow_statement = company.cash_flow
    operating_cash_flow = cash_flow_statement.loc["Operating Cash Flow"]
    capital_expenditure = cash_flow_statement.loc["Capital Expenditure"]

    free_cash_flow = operating_cash_flow + capital_expenditure
    print(f"Free Cash Flow")
    print((free_cash_flow).round(2))

    return free_cash_flow

analyze_company("AAPL")

```

```python
import yfinance as yf
import pandas as pd

def crash_analysis(ticker,crash_year):

    company = yf.Ticker(ticker)

    history = company.history(period="max")
    history.index = pd.to_datetime(history.index)
    history_year = history[history.index.year == crash_year]

    start_of_year_stock_price = history_year["Close"].iloc[0]
    print(f"Start of {crash_year} year stock price is : {start_of_year_stock_price:.2f}")

    lowest_close_price = min(history_year["Close"])
    print(f"Lowest closing price for the year {crash_year} is : {lowest_close_price:.2f}")


    percentage_drop = ((start_of_year_stock_price - lowest_close_price)/start_of_year_stock_price) * 100
    print(f"drop percentage: {percentage_drop:.2f}%")

    rock_bottom_date = history_year["Close"].idxmin()
    print(f"Rock Bottom Date: {rock_bottom_date}")

    same_year_candiates = history_year[(history_year["Close"] > start_of_year_stock_price) & (history_year.index > rock_bottom_date)]

    if len(same_year_candiates) > 0:

       recovery_date = same_year_candiates.index[0]

    else:
       search_year = crash_year + 1
       recovery_date = None

       while recovery_date is None:

          next_year_date = history[history.index.year == search_year]
          candidates = next_year_date[next_year_date["Close"] > start_of_year_stock_price]


          if len(candidates) > 0:
            recovery_date = candidates.index[0]
          else:
            search_year += 1

    print(f"Recovery date: {recovery_date}")

    recovery_time = recovery_date - rock_bottom_date
    print(f"Recovery time : {recovery_time}")

    print(f"Recovery price: {history.loc[recovery_date]['Close']:.2f}")

crash_analysis("MSFT",2000)
```

```python
import yfinance as yf
def DCF(ticker):
    company = yf.Ticker(ticker)
    print(company.income_stmt.loc["Total Revenue"])

    revenue = company.income_stmt.loc["Total Revenue"]
    revenue_sort = revenue.iloc[::-1]
    revenue_growth = revenue_sort.pct_change()
    print(revenue_growth)

    average_growth_rate = revenue_growth.dropna().mean()
    print(f"Average Growth Rate: {average_growth_rate:.2}")

    fcf = analyze_company(ticker)
    print(fcf)

    latest_fcf = fcf.iloc[0]
    print(f"Latest Free cash flow Price: {latest_fcf:,.2f}")

    projected_fcf = []
    fcf_value = latest_fcf

    for year in range(1,6):
        fcf_value = fcf_value * (1 + average_growth_rate)
        projected_fcf.append(fcf_value)
        print(f"Year {year} FCF: {fcf_value:,.2f}")

DCF("AAPL")
```
